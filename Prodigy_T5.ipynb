```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
accidents_df = pd.read_csv('accidents_2012_to_2014.csv')

# Display the first few rows of the dataset
print(accidents_df.head())

# Explore the data types and missing values
print(accidents_df.info())

# Summary statistics
print(accidents_df.describe())

# Data cleaning and preprocessing (if needed)
# Check for missing values
print(accidents_df.isnull().sum())

# Analyze patterns related to road conditions, weather, and time of day
# Road conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_Surface_Conditions', data=accidents_df, palette='viridis')
plt.title('Distribution of Road Surface Conditions')
plt.xlabel('Road Surface Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_Conditions', data=accidents_df, palette='viridis')
plt.title('Distribution of Weather Conditions')
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Time of day
accidents_df['Time'] = pd.to_datetime(accidents_df['Time'])
accidents_df['Hour'] = accidents_df['Time'].dt.hour
plt.figure(figsize=(10, 6))
sns.histplot(accidents_df['Hour'], bins=24, kde=False, color='skyblue', edgecolor='black')
plt.title('Distribution of Accidents Over Hours of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')
plt.xticks(range(24))
plt.grid(True)
plt.show()

# Visualize accident hotspots
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Longitude', y='Latitude', data=accidents_df, alpha=0.5)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()

# Contributing factors
plt.figure(figsize=(10, 6))
sns.countplot(x='Number_of_Vehicles', data=accidents_df, palette='viridis')
plt.title('Distribution of Number of Vehicles Involved in Accidents')
plt.xlabel('Number of Vehicles')
plt.ylabel('Count')
plt.show()
```

```python
# Road type
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_Type', data=accidents_df, palette='viridis')
plt.title('Distribution of Road Types')
plt.xlabel('Road Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Junction detail
plt.figure(figsize=(10, 6))
sns.countplot(x='Junction_Detail', data=accidents_df, palette='viridis')
plt.title('Distribution of Junction Details')
plt.xlabel('Junction Detail')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Light conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Light_Conditions', data=accidents_df, palette='viridis')
plt.title('Distribution of Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Special conditions at site
plt.figure(figsize=(10, 6))
sns.countplot(x='Special_Conditions_at_Site', data=accidents_df, palette='viridis')
plt.title('Distribution of Special Conditions at Site')
plt.xlabel('Special Conditions at Site')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Urban or rural area
plt.figure(figsize=(10, 6))
sns.countplot(x='Urban_or_Rural_Area', data=accidents_df, palette='viridis')
plt.title('Distribution of Urban or Rural Areas')
plt.xlabel('Urban or Rural Area')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
